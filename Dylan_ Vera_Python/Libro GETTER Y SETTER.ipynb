```python
class libro:
    #Vamos a aplicar de la forma tradicional
    #Encapsulamiento y validaciones seguras a los atributos de la clase libro
    def __init__(self, titulo, precio):
        self._titulo = titulo
        self._precio = precio

    # GETTER = CONSULTAR, VER, MOSTRAR: NOS PERMITE ACCEDER AL VALOR ALMACENADO EN EL ATRIBUTO DE FORMA SEGURA

    def get_titulo(self):
        #vamos a devolver el valor almacenado en el titulo
        return self._titulo

    def get_precio(self):
        return self._precio

    # SETTER = MODIFICAR LOS VALORES, CONTROLAR Y VALIDAR LOS VALORES ANTES DE SER MODIFICADOS.
    def set_titulo(self, nuevo_titulo):
        if isinstance(nuevo_titulo, str):
            self._titulo = nuevo_titulo
        else:
            print("Debe ingresar un titulo valido")

    def set_precio(self, nuevo_precio):
        if isinstance(nuevo_precio, (int, float)) and nuevo_precio >= 0:
            self._precio = nuevo_precio
        else:
            print("Debe ingresar un precio con un numero valido\n")

    # Metodo adicional para mostrar la informacion completa del libro
    def mostrar_info(self):
        print(f"Titulo: {self._titulo}")
        print(f"Precio: ${self._precio:,.2f}")


# El metodo principal, experiencia de usuario
def main():
    print("\n=== SISTEMA DE LIBROS ADSO ===")


# Creacion del objeto
libro1 = libro("El Quijote",80000)

# Mostraremos la informacion principal osea los getter
print("Informacion principal del libro \n")
libro1.mostrar_info()


# Actualizacion del precio
print("Precio actual del libro\n")
print(f"Precio actual:", libro1.get_precio())

# Quiero que me salga un error en la actualizacion del typado del precio del libro
libro1.set_precio("Hola")
# Actualizaremos el precio de forma correcta
libro1.set_precio(90000)
print(f"El nuevo precio actualizado es: ", libro1.get_precio())

print("Informacio principal nueva del libro\n")
libro1.mostrar_info()

print("======= PROGRAMA FINALIZADO =======")

if __name__ == "__main__":
    main()
```

```python
class estudiante:
    # Aplicamos el encapsulamiento y validaciones seguras a la clase Estudiante

    def __init__(self, nombre, documento, edad, programa, promedio):
          self.__nombre = nombre
          self.__documento = documento
          self.__edad = edad
          self.__programa = programa
          self.__promedio = promedio


    # ======= GETTERS =======

    def get__nombre(self):
          return self.__nombre


    def get__documento(self):
          return self.__documento


    def get_edad(self):
          return self.__edad


    def get__programa(self):
          return self.__programa


    def get__promedio(self):
          return self.__promedio


    # ======= SETTERS =======

    def set__nombre(self, nuevo_nombre):
          if isinstance(nuevo_nombre, str) and nuevo_nombre.strip():
                self.__nombre = nuevo_nombre
          else:
                print("Debe ingresar un nombre valido")


    def set__documento(self, nuevo_documento):
          if isinstance(nuevo_documento, str) and nuevo_documento.strip():
                self.__documento = nuevo_documento
          else:
                print("Debe ingresar un documento valido")


    def set__edad(self, nueva_edad):
          if isinstance(nueva_edad, int) and nueva_edad > 0:
                self.__edad = nueva_edad
          else:
                print("Debe ingresar una edad valida")


    def set__programa(self, nuevo_programa):
          if isinstance(nuevo_programa, str) and nuevo_programa.strip():
                self.__programa = nuevo_programa
          else:
                print("Debe ingresar un programa valido")


    def set__promedio(self, nuevo_promedio):
          if isinstance(nuevo_promedio, (int, float)) and 0 <= nuevo_promedio <= 5:
                self.__promedio = nuevo_promedio
          else:
                print("Debe ingresar un promedio valido")


    # ======= METODO PARA MOSTRAR INFO =======

    def mostrar__info(self):
          print("Nombre:", self.__nombre)
          print("Documento:", self.__documento)
          print("Edad:", self.__edad)
          print("Programa:", self.__programa)
          print("Promedio:", self.__promedio)


# ======= METODO PRINCIPAL =======

def main():
      print("\n=== SISTEMA DE ESTUDIANTES ADSO ===")

      # ======= CREACION DEL OBJETO =======
      estudiante1 = estudiante("Juan Perez", "123456", 20, "ADSO", 4.2)

      print("\nInformacion actual del estudiante\n")
      estudiante1.mostrar__info()

      print("\nPromedio actual del estudiante")
      print("promedio:", estudiante1.get__promedio())

      # ======= ACTUALIZAR PROMEDIO =======
      estudiante1.set__promedio(4.8)

      print("\nNuevo promedio actualizado:", estudiante1.get__promedio())

      print("\nInformacion actualizada del estudiante\n")
      estudiante1.mostrar__info()

      print("\n======= PROGRAMA FINALIZADO =======")


if __name__== "__main__":
      main()
```

```python
class Empleado:
    def trabajar(self):
          print("El empleado esta realizando tareas generales")

class Gerente(Empleado):
    def trabajar(self):
          print("El Gerente administra la empresa")
    pass

class Vendedor(Empleado):
    def trabajar(self):
          print("El Vendedor atiende un cliente")
    pass

def main():
    gerente1 = Gerente()
    gerente1.trabajar()
    vendedor1 = Vendedor()
    vendedor1.trabajar()


if __name__ == "__main__":
    main()
```

```python
class Padre:
    def __init__(self, mensaje): # Constructor, creo un unico atributo para ser heredado
        self.mensaje = mensaje

class Hijo (Padre):
    def __init__(self, mensaje, nombre):
      super().__init__(mensaje) # Aqui invoco el atributo que esta en el PADRE
      self.nombre = nombre

def main():
      objepadre = Padre("Este es el mensaje desde el padre")
      print (objepadre.mensaje)
      objetohijo = Hijo("Este es el mensaje enviado desde Hijo","JOSE PEREZ")
      print (objetohijo.mensaje)
      print (objetohijo.nombre)


if __name__ == "__main__":
      main()
```

```python
class Operacion:
    def __init__(self, a ,b): # Se genera el constructor
        self.a = a # Se crea el primer atributo
        self.b = b # Se crea el segundo atributo
        print("Inicializando Operación...")
    def calcular(self):
        print("Operación Generica")
class Suma(Operacion):
    def __init__(self, a, b):
        super().__init__(a,b)
        print("Preparando operacion SUMA")
    def calcular(self):
        super().calcular() # Aqui llamamos al metodo del padre
        resultado = self.a + self.b # Se asigna el signo a la operacion correspondiente
        print(f"La SUMA es:{resultado}")

class Resta(Operacion):
    def __init__(self, a, b):
        super().__init__(a,b)
        print("Preparando operacion RESTA")
    def calcular(self):
        super().calcular() # Aqui llamamos al metodo del padre
        resultado = self.a - self.b # Se asigna el signo a la operacion correspondiente
        print(f"La RESTA es:{resultado}")

class Multiplicacion(Operacion):
    def __init__(self, a, b):
        super().__init__(a,b)
        print("Preparando operacion MULTIPLICACION")
    def calcular(self):
        super().calcular() # Aqui llamamos al metodo del padre
        resultado = self.a * self.b # Se asigna el numero a la operacion corespondiente
        print(f"La NULTIPLICACION es:{resultado}")

class Division(Operacion):
    def __init__(self, a, b):
        super().__init__(a, b)
        print("Preparando operacion division")

    def calcular(self):
        super().calcular() # Aqui se llama al metodo padre
        if self.b == 0:    # Se invoca un if para evitar la division por 0
            raise ValueError("Error: No se puede dividir entre cero")
        else:
         print("La division es:", self.a / self.b) # Se conecta el signo correspondiente a la operacion

class Potenciacion(Operacion):
    def __init__(self, a, b):
        super().__init__(a, b) # Aqui se llama al metodo padre
        print("Preparando operacion potencia")

    def calcular(self):
        super().calcular()

        if self.b > 3: # Se invoca un if para evitar una potenciacion mayor a 3 y menor o igual a 0
            raise ValueError("Error: La potencia no puede ser mayor que 3")
        else:
         print("La potencia es:", self.a ** self.b) # Se conecta el signo correspondiente a la operacion


def main():
    num1 = float(input("Ingrese el primer numero:")) # Se le solicita al usuario asignarle un valor numerico a el dato num1
    num2 = float(input("Ingrese el segundo numero:")) # Se le solicita al usuario asignarle un valor numerico al el dato num2
    operacion1 = Suma(num1,num2) # Se indica la operacion que se ejecutara con los datos recibidos
    operacion1.calcular()
    print("----------------------------------------------------------------")

    num1 = float(input("Ingrese el primer numero:")) # Se le solicita al usuario asignarle un valor numerico a el dato num1
    num2 = float(input("Ingrese el segundo numero:")) # Se le solicita al usuario asignarle un valor numerico al el dato num2
    operacion1 = Resta(num1,num2) # Se indica la operacion que se ejecutara con los datos recibidos
    operacion1.calcular()
    print("----------------------------------------------------------------")

    num1 = float(input("Ingrese el primer numero:")) # Se le solicita al usuario asignarle un valor numerico a el dato num1
    num2 = float(input("Ingrese el segundo numero:")) # Se le solicita al usuario asignarle un valor numerico al el dato num2
    operacion1 = Multiplicacion(num1,num2) # Se indica la operacion que se ejecutara con los datos recibidos
    operacion1.calcular()
    print("----------------------------------------------------------------")

    num1 = float(input("Ingrese el primer numero:")) # Se le solicita al usuario asignarle un valor numerico a el dato num1
    num2 = float(input("Ingrese el segundo numero:")) # Se le solicita al usuario asignarle un valor numerico al el dato num2
    operacion1 = Division(num1,num2) # Se indica la operacion que se ejecutara con los datos recibidos
    operacion1.calcular()
    print("----------------------------------------------------------------")

    num1 = float(input("Ingrese el primer numero:")) # Se le solicita al usuario asignarle un valor numerico a el dato num1
    num2 = float(input("Ingrese el segundo numero:")) # Se le solicita al usuario asignarle un valor numerico al el dato num2
    operacion1 = Potenciacion(num1,num2) # Se indica la operacion que se ejecutara con los datos recibidos
    operacion1.calcular()
    print("----------------------------------------------------------------")

if __name__ == "__main__": # Se cierra el code y se ejecuta
  main()
```

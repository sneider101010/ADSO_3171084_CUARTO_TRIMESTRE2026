```python
from abc import ABC, abstractmethod
    #Crearemos la clase abstracta
class Figura(ABC):
    @abstractmethod
    def calcular_area(self):
      pass



    #Crearemos una subclase
class Circulo(Figura):
    def __init__ (self,radio):
     self.radio = radio

    def calcular_area(self):
        return 3.1416 * (self.radio**2)
#********************************************************************************


class Triangulo(Figura):
    def __init__ (self,base,altura):
     self.base = base
     self.altura = altura

    def calcular_area(self):
        return self.base * self.altura/2
class Rectangulo(Figura):
    def __init__ (self,base,altura):
     self.base = base
     self.altura = altura

    def calcular_area(self):
        return self.base * self.altura
class Cubo(Figura):
    def __init__ (self,lado):
     self.lado = lado

    def calcular_area(self):
        return 6 * self.lado ** 3
def mostrar_area(figura: Figura):
    print(f"Area: {figura.calcular_area():,.2f}")


#Ejecucion del programa


def main():
    circulo1 = Circulo(5)
    mostrar_area(circulo1)

    triangulo1 = Triangulo(5,10)
    mostrar_area(triangulo1)

    rectangulo1 = Rectangulo(5,10)
    mostrar_area(rectangulo1)


if __name__ == "__main__":
    main()
```

```python
from abc import ABC, abstractmethod
# clase ABSTRACTA  con encapsulamiento y gestion segura
class Empleado(ABC):
    def __init__(self, nombre: str) -> None:
# Vamos a almacenar como privado
      self._nombre = None
      self._nombre = nombre # Automaticamente se llama al setter
#GETTER
    @property
    def nombre(self) -> str:
  # Nombre del empleado
      return self._nombre
    @nombre.setter  # SETTER profesional con validacion basica
    def nombre(self, nombre, str) -> None:
        if isinstance(nombre,str) and nombre.strip():
            self._nombre = nombre.strp()
        else:
            raise ValueError("El nombre debe ser un texto valido")

    @abstractmethod
    def trabajar (self) -> None:
        pass

# Ahora carece de los hijos
class Gerente(Empleado):
    def trabajar(self) -> None:
      print(f"{self.nombre} esta realizando administracion de empresas.")
class Desarrolador(Empleado):
    def trabajar(self) -> None:
      print(f"{self.nombre} esta programando los diagramas de clase en poo java.")
def ejecutar_tarea(empleado: Empleado) -> None:
    empleado.trabajar()
def main():
    gerente1 = Gerente("Laura Perez")
    desarrolador1 = Desarrolador ("Kevin bareto")
    ejecutar_tarea(gerente1)
    ejecutar_tarea(desarrolador1)
    print("Actualizando nombre de desarrollador")
    desarrolador1._nombre = "Dylan Vera"
    print(f"El nuevo programador es: {desarrolador1.nombre}")
    print("n/*********** FIN DEL PROGRAMA EMPLEADOS /**********")
if __name__ == "__main__":
    main()
```

```python
from abc import ABC, abstractmethod
# clase ABSTRACTA  con encapsulamiento y gestion segura
class PersonalMedico(ABC):
    def __init__(self, nombre: str) -> None:
# Vamos a almacenar como privado
      self._nombre = None
      self._nombre = nombre # Automaticamente se llama al setter
#GETTER
    @property
    def nombre(self) -> str:
  # Nombre del empleado en el personal medico
      return self._nombre
    @nombre.setter  # SETTER profesional con validacion basica
    def nombre(self, nombre, str) -> None:
        if isinstance(nombre,str) and nombre.strip():
            self._nombre = nombre.strp()
        else:
            raise ValueError("El cargo debe ser un texto valido")

    @abstractmethod
    def trabajar(self) -> None:
        pass

# Ahora carece de los hijos
class Doctor(PersonalMedico):
    def trabajar(self) -> None:
      print(f"{self.nombre} está atendiendo pacientes en consulta.")
class Enfermero(PersonalMedico):
    def trabajar(self) -> None:
      print(f"{self.nombre} está asistiendo a los pacientes.")
class Cirujano(PersonalMedico):
    def trabajar(self) -> None:
      print(f"{self.nombre} está realizando una cirugía.")
def ejecutar_tarea(personalmedico: PersonalMedico) -> None:
    personalmedico.trabajar()
def main():
    doctor1 = Doctor ("Raul Castiblanco")
    enfermero1 = Enfermero ("Pepe Durango")
    cirujano1 = Cirujano ("Santiago Hernandez")

    ejecutar_tarea(doctor1)
    ejecutar_tarea(enfermero1)
    ejecutar_tarea(cirujano1)
    print("n**********\
     FIN DEL PROGRAMA PERSONAL MEDICO /**********")
if __name__ == "__main__":
    main()
```

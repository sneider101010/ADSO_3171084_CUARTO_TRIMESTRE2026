```python
#Herencia

class Vehiculo:  # clase padre
   def mover(self): # metodo
      print("El vehiculo se esta moviendo")


class Carro(Vehiculo): # clase hija
   pass # ni atributos ni metodos (vacia)


#===============================================================


def main(): # experiencia de usuario
   print("CLASE PADRE Y SU METODO\n")
   Vehiculo1 = Vehiculo()
   Vehiculo1.mover() #invoca el metodo mover() de la clase padre


   print("Utilizaremos el metodo mover() de la clase padre en la clase HIJO\n")
   carro1 = Carro()
   carro1.mover()


   if __name__ == "__main__":
      main()
```

```python
class Aprendiz:

    def __init__(self, nombre, apellido, id_aprendiz, estatura, edad, trimestre):
        self.__nombre = nombre
        self.__apellido = apellido
        self.__id = id_aprendiz
        self.__estatura = estatura
        self.__edad = edad
        self.__trimestre = trimestre

    def get_estatura(self):
        return self.__estatura

    def get_edad(self):
        return self.__edad

    def get_trimestre(self):
        return self.__trimestre

    def set_estatura(self, nueva_estatura):
        if isinstance(nueva_estatura, (int, float)) and nueva_estatura > 0:
            self.__estatura = nueva_estatura

    def set_edad(self, nueva_edad):
        if isinstance(nueva_edad, int) and nueva_edad > 0:
            self.__edad = nueva_edad

    def set_trimestre(self, nuevo_trimestre):
        if isinstance(nuevo_trimestre, int) and nuevo_trimestre > 0:
            self.__trimestre = nuevo_trimestre

    def Mostrar_todo(self):
        print("\n----- DATOS DEL APRENDIZ -----")
        print("Nombre:", self.__nombre)
        print("Apellido:", self.__apellido)
        print("ID:", self.__id)
        print("Estatura:", self.__estatura)
        print("Edad:", self.__edad)
        print("Trimestre:", self.__trimestre)
        print("------------------------------")


aprendiz1 = Aprendiz("Camilo", "Hurtado", "12345", 1.70, 14, 1)

aprendiz1.Mostrar_todo()

nueva_estatura = float(input("Nueva estatura: "))
aprendiz1.set_estatura(nueva_estatura)

nueva_edad = int(input("Nueva edad: "))
aprendiz1.set_edad(nueva_edad)

nuevo_trimestre = int(input("Nuevo trimestre: "))
aprendiz1.set_trimestre(nuevo_trimestre)

aprendiz1.Mostrar_todo()
```

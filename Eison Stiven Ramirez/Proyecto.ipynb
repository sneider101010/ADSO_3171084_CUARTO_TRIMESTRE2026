```python
def agregar_cliente(lista_clientes, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_clientes.append(nombre.title())
        print(f"Cliente agregado {nombre.title()}")
    else:
        print("Debe agregar un nombre válido entre 2 y 50 caracteres")


def mostrar_clientes(lista_clientes):
    print("----- LISTA DE CLIENTES -----")
    for cliente in lista_clientes:
        print(f"- {cliente}")
    print("-----------------------------")


def modificar_cliente(lista_clientes, indice, nuevo_nombre):
    if not (isinstance(nuevo_nombre, str) and 2 <= len(nuevo_nombre) <= 50):
        print("Nombre de cliente inválido, debe tener entre 2 y 50 caracteres")
        return

    if 0 <= indice < len(lista_clientes):
        original = lista_clientes[indice]
        lista_clientes[indice] = nuevo_nombre.title()
        print(f"Cliente modificado de {original} a -> {nuevo_nombre.title()}")
    else:
        print("No existe este dato en la lista o índice incorrecto.")


def eliminar_cliente(lista_clientes, indice):
    if 0 <= indice < len(lista_clientes):
        eliminado = lista_clientes.pop(indice)
        print(f"Cliente eliminado: {eliminado}")
    else:
        print("No se puede eliminar el cliente, índice fuera de rango.")


def main():
    clientes = ["Ana", "Carlos", "Beatriz"]

    print("\n==============================")
    print("     Dylan maricon")
    print("==============================")
    mostrar_clientes(clientes)

    print("\n------------------------------")
    print("       AGREGAR CLIENTE")
    print("------------------------------")
    agregar_cliente(clientes, "Juan")
    mostrar_clientes(clientes)

    print("\n------------------------------")
    print("      MODIFICAR CLIENTE")
    print("------------------------------")
    modificar_cliente(clientes, 0, "Maria")
    mostrar_clientes(clientes)

    print("\n------------------------------")
    print("       ELIMINAR CLIENTE")
    print("------------------------------")
    eliminar_cliente(clientes, 2)
    mostrar_clientes(clientes)

    print("\n==============================")
    print("     PROCESO FINALIZADO")
    print("==============================")


if __name__ == "__main__":
    main()
```

```python
def agregar_componente(lista_componentes, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_componentes.append(nombre.title())
        print(f"Componente agregado: {nombre.title()}")
    else:
        print("Debe ingresar un nombre válido (entre 2 y 50 caracteres)")


def mostrar_componentes(lista_componentes):
    print("----- INVENTARIO DE COMPONENTES -----")
    if not lista_componentes:
        print("  (No hay componentes registrados aún)")
    for i, componente in enumerate(lista_componentes, 1):
        print(f"{i:2d}. {componente}")
    print("-------------------------------------")


def modificar_componente(lista_componentes, indice, nuevo_nombre):
    if not (isinstance(nuevo_nombre, str) and 2 <= len(nuevo_nombre) <= 50):
        print("Nombre inválido: debe tener entre 2 y 50 caracteres")
        return

    # Convertimos a índice basado en 0 (como en listas)
    indice_real = indice - 1

    if 0 <= indice_real < len(lista_componentes):
        original = lista_componentes[indice_real]
        lista_componentes[indice_real] = nuevo_nombre.title()
        print(f"Componente modificado: {original} → {nuevo_nombre.title()}")
    else:
        print("Índice fuera de rango. No se pudo modificar.")


def eliminar_componente(lista_componentes, indice):
    indice_real = indice - 1
    if 0 <= indice_real < len(lista_componentes):
        eliminado = lista_componentes.pop(indice_real)
        print(f"Componente eliminado: {eliminado}")
    else:
        print("Índice fuera de rango. No se pudo eliminar.")


def main():
    # Inventario inicial de ejemplo
    componentes = [
        "Procesador Ryzen 5 5600X",
        "Memoria 16GB DDR4 3200MHz",
        "Tarjeta gráfica RTX 3060"
    ]

    print("\n" + "="*45)
    print("   GESTIÓN DE COMPONENTES DE PC - 2025")
    print("="*45)

    mostrar_componentes(componentes)

    print("\n" + "-"*40)
    print(" AGREGAR COMPONENTE")
    print("-"*40)
    agregar_componente(componentes, "Disco SSD NVMe 1TB")
    mostrar_componentes(componentes)

    print("\n" + "-"*40)
    print(" MODIFICAR COMPONENTE")
    print("-"*40)
    # Modificamos el primero (índice 1 porque ahora mostramos desde 1)
    modificar_componente(componentes, 1, "Procesador Ryzen 7 5800X")
    mostrar_componentes(componentes)

    print("\n" + "-"*40)
    print(" ELIMINAR COMPONENTE")
    print("-"*40)
    # Eliminamos el que ahora está en posición 3
    eliminar_componente(componentes, 3)
    mostrar_componentes(componentes)

    print("\n" + "="*45)
    print("          PROCESO FINALIZADO")
    print("="*45)


if __name__ == "__main__":
    main()
```

```python
import os
import sys
from datetime import datetime

# ==================== VARIABLES GLOBALES ====================

inventario = {
    "001": {"nombre": "Martillo", "precio": 150000, "stock": 25},
    "002": {"nombre": "Destornillador plano", "precio": 45000, "stock": 40},
    "003": {"nombre": "Taladro eléctrico", "precio": 850000, "stock": 8},
    "004": {"nombre": "Llave inglesa 12\"", "precio": 120000, "stock": 18},
    "005": {"nombre": "Sierra manual", "precio": 95000, "stock": 12},
    "006": {"nombre": "Cinta métrica 5m", "precio": 35000, "stock": 35},
    "007": {"nombre": "Nivel de burbuja", "precio": 65000, "stock": 22},
    "008": {"nombre": "Juego de destornilladores (6 piezas)", "precio": 180000, "stock": 15},
    "009": {"nombre": "Broca para concreto 8mm", "precio": 25000, "stock": 80},
    "010": {"nombre": "Tornillos para madera 1\" (caja 100)", "precio": 18000, "stock": 45},
    "011": {"nombre": "Pintura acrílica blanca 1 galón", "precio": 125000, "stock": 10}
}

factura_actual = []
historial_ventas = []          # Guarda todas las ventas completadas
ventas_del_dia = 0             # Contador de ventas realizadas hoy
total_ventas_dia = 0.0         # Monto total vendido hoy
fecha_hoy = datetime.now().strftime("%Y-%m-%d")

def limpiar_pantalla():
    os.system("cls" if os.name == "nt" else "clear")

def formato_precio(precio):
    """Formatea el precio con separador de miles y símbolo $"""
    return f"${precio:,.0f}"
  |
def mostrar_encabezado_tabla():
    print(f"{'ID':<6}{'NOMBRE':<30}{'PRECIO':<15}{'STOCK':<8}")
    print("-" * 65)

def mostrar_inventario():
    print("\n" + "═" * 70)
    print("              INVENTARIO COMPLETO - FERRETERÍA 'LE FALTA UN TORNILLO'")
    print("═" * 70)
    if not inventario:
        print("El inventario está vacío.")
    else:
        mostrar_encabezado_tabla()
        for id_prod, datos in inventario.items():
            print(f"{id_prod:<6}{datos['nombre']:<30}{formato_precio(datos['precio']):<15}{datos['stock']:<8}")
    print("-" * 65)

def buscar_producto_nombre():
    print("\n****** BUSCAR PRODUCTO POR NOMBRE ******")
    termino = input("Ingrese el nombre o parte del nombre: ").strip()
    encontrado = False
    print("\nResultado de búsqueda:")
    mostrar_encabezado_tabla()
    for id_prod, datos in inventario.items():
        if termino.lower() in datos["nombre"].lower():
            print(f"{id_prod:<6}{datos['nombre']:<30}{formato_precio(datos['precio']):<15}{datos['stock']:<8}")
            encontrado = True
    if not encontrado:
        print("No se encontraron productos que coincidan con la búsqueda.")
    print("-" * 65)

def agregar_nuevo_producto():
    print("\n****** AGREGAR NUEVO PRODUCTO ******")
    id_prod = input("Ingrese el ID del nuevo producto: ").strip()
    if id_prod in inventario:
        print(f"❌ El producto con ID {id_prod} ya existe.")
        return

    nombre = input("Nombre del producto: ").strip()
    try:
        precio = float(input("Precio del producto: "))
        stock = int(input("Stock inicial: "))
        if precio < 0 or stock < 0:
            print("❌ El precio y el stock no pueden ser negativos.")
            return

        inventario[id_prod] = {"nombre": nombre, "precio": precio, "stock": stock}
        print(f"✅ Producto '{nombre}' agregado correctamente con ID {id_prod}.")
    except ValueError:
        print("❌ Error: Ingrese valores numéricos válidos.")

def modificar_producto():
    print("\n****** MODIFICAR PRODUCTO ******")
    mostrar_inventario()
    id_prod = input("\nIngrese el ID del producto a modificar: ").strip()

    if id_prod not in inventario:
        print("❌ Producto no encontrado.")
        return

    prod = inventario[id_prod]
    print(f"\nProducto actual: {prod['nombre']} | Stock: {prod['stock']} | Precio: {formato_precio(prod['precio'])}")

    print("\n¿Qué desea modificar?")
    print("1. Agregar unidades al stock")
    print("2. Cambiar precio")
    print("3. Cambiar nombre")
    print("4. Cancelar")
    op = input("Seleccione una opción: ").strip()

    try:
        if op == "1":
            cantidad = int(input("¿Cuántas unidades desea agregar? "))
            if cantidad > 0:
                prod['stock'] += cantidad
                print(f"✅ Stock actualizado: {prod['stock']} unidades.")
            else:
                print("La cantidad debe ser mayor a 0.")

        elif op == "2":
            nuevo_precio = float(input("Ingrese el nuevo precio: "))
            if nuevo_precio >= 0:
                prod['precio'] = nuevo_precio
                print(f"✅ Precio actualizado a {formato_precio(nuevo_precio)}")
            else:
                print("El precio no puede ser negativo.")

        elif op == "3":
            nuevo_nombre = input("Ingrese el nuevo nombre: ").strip()
            if nuevo_nombre:
                prod['nombre'] = nuevo_nombre
                print(f"✅ Nombre actualizado a: {nuevo_nombre}")
            else:
                print("El nombre no puede estar vacío.")

        elif op == "4":
            print("Operación cancelada.")
        else:
            print("Opción inválida.")
    except ValueError:
        print("❌ Error: Ingrese datos numéricos válidos.")

def eliminar_producto():
    print("\n****** ELIMINAR PRODUCTO ******")
    mostrar_inventario()
    id_prod = input("\nIngrese el ID del producto a eliminar: ").strip()

    if id_prod in inventario:
        prod = inventario[id_prod]
        print(f"Va a eliminar: {prod['nombre']} (Stock: {prod['stock']})")
        confirmacion = input("¿Está seguro? (s/n): ").lower()
        if confirmacion == "s":
            eliminado = inventario.pop(id_prod)
            print(f"✅ Producto '{eliminado['nombre']}' eliminado correctamente.")
        else:
            print("Operación cancelada.")
    else:
        print("❌ ID no encontrado.")

def vender_producto():
    global ventas_del_dia, total_ventas_dia

    print("\n****** VENTA DE PRODUCTO ******")
    mostrar_inventario()

    if input("\n¿Desea buscar por nombre primero? (s/n): ").lower() == "s":
        buscar_producto_nombre()

    id_prod = input("\nIngrese el ID del producto a vender: ").strip()

    if id_prod in inventario:
        prod = inventario[id_prod]
        print(f"\nProducto seleccionado: {prod['nombre']} | Stock disponible: {prod['stock']} | Precio: {formato_precio(prod['precio'])}")

        try:
            cantidad = int(input("Cantidad a vender: "))
            if 0 < cantidad <= prod['stock']:
                subtotal = cantidad * prod['precio']
                prod['stock'] -= cantidad

                factura_actual.append({
                    "id": id_prod,
                    "nombre": prod['nombre'],
                    "cantidad": cantidad,
                    "precio": prod['precio'],
                    "subtotal": subtotal
                })

                print(f"✅ Agregado al carrito: {cantidad} x {prod['nombre']} → {formato_precio(subtotal)}")
            else:
                print("❌ Cantidad no válida o stock insuficiente.")
        except ValueError:
            print("❌ Error: Ingrese un número válido.")
    else:
        print("❌ Producto no encontrado.")

def generar_factura():
    global ventas_del_dia, total_ventas_dia

    print("\n" + "═" * 50)
    print("       FERRETERÍA 'LE FALTA UN TORNILLO'")
    print("               FACTURA DE VENTA")
    print("═" * 50)
    print(f"Fecha: {datetime.now().strftime('%d/%m/%Y %H:%M:%S')}")
    print("-" * 50)

    if not factura_actual:
        print("No hay ítems en la factura actual.")
        return

    total = 0
    print(f"{'CANT':<5}{'ARTÍCULO':<28}{'PRECIO':<12}{'SUBTOTAL':<12}")
    print("-" * 60)

    for item in factura_actual:
        print(f"{item['cantidad']:<5}{item['nombre']:<28}{formato_precio(item['precio']):<12}{formato_precio(item['subtotal']):<12}")
        total += item['subtotal']

    print("-" * 60)
    print(f"{'TOTAL A PAGAR:':<45}{formato_precio(total)}")
    print("═" * 50)

    if input("\n¿Confirma el pago y cerrar la venta? (s/n): ").lower() == "s":
        # Guardar en historial
        venta_completa = {
            "fecha": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "items": factura_actual.copy(),
            "total": total
        }
        historial_ventas.append(venta_completa)

        ventas_del_dia += 1
        total_ventas_dia += total

        factura_actual.clear()
        print("\n✅ Venta finalizada exitosamente. ¡Gracias por su compra!")
    else:
        print("Venta cancelada.")

def mostrar_reporte_ventas():
    print("\n" + "═" * 60)
    print("               REPORTE DE VENTAS DEL DÍA")
    print("═" * 60)
    print(f"Fecha: {datetime.now().strftime('%d/%m/%Y')}")
    print(f"Ventas realizadas hoy: {ventas_del_dia}")
    print(f"Total vendido hoy: {formato_precio(total_ventas_dia)}")
    print("-" * 60)

    if not historial_ventas:
        print("Aún no se han registrado ventas.")
    else:
        print(f"Últimas {min(5, len(historial_ventas))} ventas:")
        for i, venta in enumerate(reversed(historial_ventas[-5:]), 1):
            print(f"{i}. {venta['fecha']} → {formato_precio(venta['total'])}")

def main():
    global fecha_hoy

    while True:
        limpiar_pantalla()
        print("═" * 85)
        print("       SISTEMA DE GESTIÓN DE FERRETERÍA 'LE FALTA UN TORNILLO'      ")
        print("═" * 85)
        print(f"Fecha: {datetime.now().strftime('%d/%m/%Y')} | Ventas hoy: {ventas_del_dia} | Total: {formato_precio(total_ventas_dia)}")
        print("-" * 85)

        print("1. Mostrar Inventario Completo")
        print("2. Buscar producto por nombre")
        print("3. Vender producto (agregar a factura)")
        print("4. Generar / Cerrar factura")
        print("5. Agregar nuevo producto")
        print("6. Modificar producto")
        print("7. Eliminar producto")
        print("8. Ver reporte de ventas del día")
        print("9. Salir")

        opcion = input("\nSeleccione una opción: ").strip()

        match opcion:
            case "1": mostrar_inventario()
            case "2": buscar_producto_nombre()
            case "3": vender_producto()
            case "4": generar_factura()
            case "5": agregar_nuevo_producto()
            case "6": modificar_producto()
            case "7": eliminar_producto()
            case "8": mostrar_reporte_ventas()
            case "9":
                print("\n¡Gracias por usar el sistema! Hasta pronto.")
                break
            case _:
                print("❌ Opción no válida.")

        if opcion != "9":
            input("\nPresione Enter para continuar...")

if __name__ == "__main__":
    if sys.version_info < (3, 10):
        print("Error: Esta aplicación requiere Python 3.10 o superior.")
    else:
        main()
```

```python
class Operacion:
    def __init__(self, a, b):
        self.a = a
        self.b = b
        print("Inicializando Operacion")

    def calcular(self):
        print("Operaciones genericas en matematicas")


class Suma(Operacion):
    def __init__(self, a, b):
        super().__init__(a, b)
        print("Preparando operacion suma")

    def calcular(self):
        super().calcular()
        print("La suma es:", self.a + self.b)


class Resta(Operacion):
    def __init__(self, a, b):
        super().__init__(a, b)
        print("Preparando operacion resta")

    def calcular(self):
        super().calcular()
        print("La resta es:", self.a - self.b)


class Multiplicacion(Operacion):
    def __init__(self, a, b):
        super().__init__(a, b)
        print("Preparando operacion multiplicacion")

    def calcular(self):
        super().calcular()
        print("La multiplicacion es:", self.a * self.b)


class Division(Operacion):
    def __init__(self, a, b):
        super().__init__(a, b)
        print("Preparando operacion division")

    def calcular(self):
        super().calcular()

        if self.b == 0:
            raise ValueError("Error: No se puede dividir entre cero")
        else:
            print("La division es:", self.a / self.b)

class Potencia(Operacion):
    def __init__(self, a, b):
        super().__init__(a, b)
        print("Preparando operacion potencia")

    def calcular(self):
        super().calcular()

        if self.b > 3:
            raise ValueError("Error: La potencia no puede ser mayor que 3")
        else:
            print("La potencia es:", self.a ** self.b)


def main():

    print("\n--- SUMA ---")
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
    Suma(a, b).calcular()

    print("\n--- RESTA ---")
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
    Resta(a, b).calcular()

    print("\n--- MULTIPLICACION ---")
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
    Multiplicacion(a, b).calcular()

    print("\n--- DIVISION ---")
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
    Division(a, b).calcular()

    print("\n--- POTENCIA ---")
    a = float(input("Ingrese el primer numero: "))
    b = float(input("Ingrese el segundo numero: "))
    Potencia(a, b).calcular()


    print("\n=================== FIN DEL PROGRAMA ===================")



if __name__ == "__main__":
    main()
```

```python
from abc import ABC, abstractmethod

class Figura(ABC):
    @abstractmethod
    def calcular(self): pass

# =========================================================================
class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def calcular(self):
        return 3.1416 * (self.radio ** 2)

# =========================================================================
class Triangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular(self):
        return self.base * self.altura / 2

# =========================================================================
class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular(self):
        return self.base * self.altura

# =========================================================================
class Cubo(Figura):
    def __init__(self, lado):
        self.lado = lado

    # volumen del cubo
    def calcular(self):
        return self.lado ** 3

    # area del cubo
    def calcular_area(self):
        return 6 * (self.lado ** 2)

# =========================================================================
class Cilindro(Figura):
    def __init__(self, radio, altura):
        self.radio = radio
        self.altura = altura

    # volumen del cilindro
    def calcular(self):
        return 3.1416 * (self.radio ** 2) * self.altura

    # area del cilindro
    def calcular_area(self):
        return 2 * 3.1416 * self.radio * (self.radio + self.altura)

# =========================================================================

def mostrar_resultado(figura: Figura, nombre_figura: str):
    print(f"El resultado del {nombre_figura} es: {figura.calcular():,.2f}")

# Ejecucion del programa
def main():
    circulo1 = Circulo(5)
    mostrar_resultado(circulo1, "area del circulo")

    triangulo1 = Triangulo(5, 10)
    mostrar_resultado(triangulo1, "area del triangulo")

    rectangulo1 = Rectangulo(5, 10)
    mostrar_resultado(rectangulo1, "area del rectangulo")

    cubo1 = Cubo(5)
    mostrar_resultado(cubo1, "volumen del cubo")
    print(f"El area del cubo es: {cubo1.calcular_area():,.2f}")

    cilindro1 = Cilindro(3, 7)
    mostrar_resultado(cilindro1, "volumen del cilindro")
    print(f"El area del cilindro es: {cilindro1.calcular_area():,.2f}")

    print("\n=================== FIN DEL PROGRAMA ===================")



if __name__ == "__main__":
    main()
```

```python
from abc import ABC, abstractmethod

# clase ABSTRACTA con encapsulamiento y gestion segura
class Empleado(ABC):

    def __init__(self, nombre: str) -> None:
      # vamos a almacenar como privado
      self._nombre = None
      self.nombre = nombre # usa el setter para inicializar

    # GETTER
    @property
    def nombre(self) -> str:
     # nombre del empleado
     return self._nombre

    @nombre.setter # setter profesional con validacion basica
    def nombre(self, nombre: str) -> None:
        if isinstance(nombre, str) and nombre.strip():
            self._nombre = nombre.strip()
        else:
            raise ValueError("El nombre debe ser un texto valido")


    @abstractmethod
    def trabajar(self) -> None:
        pass

# ahora crearemos a los hijos (fuera de la clase Empleado)
class Gerente(Empleado):
  def trabajar(self) -> None:
    print(f"{self.nombre} Esta realizando administracion de la empresa.")

class Desarrollador(Empleado):
  def trabajar(self) -> None:
    print(f"{self.nombre} Esta programando los diagramas en poo java.")

# polimorfismo en accion (fuera de la clase Empleado)
def ejecutar_tarea(empleado: Empleado) -> None:
   empleado.trabajar()

# crearemos la ejecucion (fuera de la clase Empleado)
def main():
   gerente1 = Gerente("Juan perez")
   desarrollador1 = Desarrollador("José Perez")
   ejecutar_tarea(gerente1)
   ejecutar_tarea(desarrollador1)
   print("Actualizacion del nombre del desarrollador")
   desarrollador1.nombre = "Kevin Turizo"
   print("El nombre del desarrollador es:", desarrollador1.nombre)
   print("\n===================FIN DEL PROGRAMA===================")

if __name__ == "__main__":
    main()
```

```python
from abc import ABC, abstractmethod
class PersonalMedico(ABC):

    def __init__(self, nombre: str) -> None:
        self._nombre = None
        self.nombre = nombre

    # GETTER
    @property
    def nombre(self) -> str:
        return self._nombre

    # SETTER
    @nombre.setter
    def nombre(self, nombre: str) -> None:
        if isinstance(nombre, str) and nombre.strip():
            self._nombre = nombre.strip()
        else:
            raise ValueError("El nombre debe ser un texto valido")

    # METODO ABSTRACTO
    @abstractmethod
    def trabajar(self):
        pass


# ================= CLASES HIJAS =================

class Doctor(PersonalMedico):
    def trabajar(self):
        return "El doctor está atendiendo pacientes en consulta."


class Enfermero(PersonalMedico):
    def trabajar(self):
        return "El enfermero está asistiendo a los pacientes."


class Cirujano(PersonalMedico):
    def trabajar(self):
        return "El cirujano está realizando una cirugía."


# ================= POLIMORFISMO =================

def mostrar_trabajo(personal: PersonalMedico):
    print(f"{personal.nombre}: {personal.trabajar()}")


# ================= EJECUCION =================

def main():
    doctor1 = Doctor("Francisco el médico")
    enfermero1 = Enfermero("José el enfermero")
    cirujano1 = Cirujano("Juan el cirujano")

    mostrar_trabajo(doctor1)
    mostrar_trabajo(enfermero1)
    mostrar_trabajo(cirujano1)

    print("\n=================== FIN DEL PROGRAMA ===================")


if __name__ == "__main__":
    main()
```

```python
def agregar_cliente(lista_clientes, nombre):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        lista_clientes.append(nombre.title())
        print(f"Cliente agregado {nombre.title()}")
    else:
        print("Debe agregar un nombre válido entre 2 y 50 caracteres")


def mostrar_clientes(lista_clientes):
    print("----- LISTA DE CLIENTES -----")
    for cliente in lista_clientes:
        print(f"- {cliente}")
    print("-----------------------------")


def modificar_cliente(lista_clientes, indice, nuevo_nombre):
    if not (isinstance(nuevo_nombre, str) and 2 <= len(nuevo_nombre) <= 50):
        print("Nombre de cliente inválido, debe tener entre 2 y 50 caracteres")
        return

    if 0 <= indice < len(lista_clientes):
        original = lista_clientes[indice]
        lista_clientes[indice] = nuevo_nombre.title()
        print(f"Cliente modificado de {original} a -> {nuevo_nombre.title()}")
    else:
        print("No existe este dato en la lista o índice incorrecto.")


def eliminar_cliente(lista_clientes, indice):
    if 0 <= indice < len(lista_clientes):
        eliminado = lista_clientes.pop(indice)
        print(f"Cliente eliminado: {eliminado}")
    else:
        print("No se puede eliminar el cliente, índice fuera de rango.")


def ordenar_clientes(lista_clientes):
    for clientes in sorted(lista_clientes):
        print(clientes)


def main():
    clientes = ["Ana", "Carlos", "Beatriz"]

    print("\n==============================")
    print("     CLIENTES ACTUALES")
    print("==============================")
    mostrar_clientes(clientes)

    print("\n------------------------------")
    print("       AGREGAR CLIENTE")
    print("------------------------------")
    agregar_cliente(clientes, "Juan")
    mostrar_clientes(clientes)

    print("\n------------------------------")
    print("      MODIFICAR CLIENTE")
    print("------------------------------")
    modificar_cliente(clientes, 0, "Maria")
    mostrar_clientes(clientes)

    print("\n------------------------------")
    print("       ELIMINAR CLIENTE")
    print("------------------------------")
    eliminar_cliente(clientes, 2)
    mostrar_clientes(clientes)

    print("\n------------------------------")
    print("       ORDENAR CLIENTE")
    print("------------------------------")
    ordenar_clientes(clientes)
    mostrar_clientes(clientes)

    print("\n==============================")
    print("     PROCESO FINALIZADO")
    print("==============================")


if __name__ == "__main__":
    main()
```

```python
def agregar_producto(lista_productos, nombre, precio):
    if isinstance(nombre, str) and 2 <= len(nombre) <= 50:
        producto = {"nombre": nombre.title(), "precio": precio}
        lista_productos.append(producto)
        print(f"Producto agregado {nombre.title()} - ${precio}")
    else:
        print("Debe agregar un nombre válido entre 2 y 50 caracteres")


def mostrar_productos(lista_productos):
    print("----- LISTA DE PRODUCTOS -----")
    for producto in lista_productos:
        print(f"- {producto['nombre']} | Precio: ${producto['precio']}")
    print("-----------------------------")


def modificar_producto(lista_productos, indice, nuevo_nombre, nuevo_precio):
    if not (isinstance(nuevo_nombre, str) and 2 <= len(nuevo_nombre) <= 50):
        print("Nombre de producto inválido")
        return

    if 0 <= indice < len(lista_productos):
        original = lista_productos[indice]
        lista_productos[indice] = {
            "nombre": nuevo_nombre.title(),
            "precio": nuevo_precio
        }
        print(f"Producto modificado de {original['nombre']} a -> {nuevo_nombre.title()}")
    else:
        print("No existe este dato en la lista o índice incorrecto.")


def eliminar_producto(lista_productos, indice):
    if 0 <= indice < len(lista_productos):
        eliminado = lista_productos.pop(indice)
        print(f"Producto eliminado: {eliminado['nombre']}")
    else:
        print("No se puede eliminar el producto, índice fuera de rango.")


def ordenar_productos(lista_productos):
    lista_productos.sort(key=lambda x: x["nombre"])
    print("Productos ordenados por nombre")


def main():
    productos = [
        {"nombre": "Sopa", "precio": 5000},
        {"nombre": "Mojarra", "precio": 12000},
        {"nombre": "Pechuga", "precio": 10000},
        {"nombre": "Lentejas", "precio": 6000}
    ]

    print("\n==============================")
    print("     PRODUCTOS ACTUALES")
    print("==============================")
    mostrar_productos(productos)

    print("\n------------------------------")
    print("       AGREGAR PRODUCTO")
    print("------------------------------")
    agregar_producto(productos, "Frijol", 7000)
    mostrar_productos(productos)

    print("\n------------------------------")
    print("      MODIFICAR PRODUCTOS")
    print("------------------------------")
    modificar_producto(productos, 0, "Pollo", 15000)
    mostrar_productos(productos)

    print("\n------------------------------")
    print("       ELIMINAR PRODUCTO")
    print("------------------------------")
    eliminar_producto(productos, 2)
    mostrar_productos(productos)

    print("\n------------------------------")
    print("       ORDENAR PRODUCTOS")
    print("------------------------------")
    ordenar_productos(productos)
    mostrar_productos(productos)

    print("\n==============================")
    print("     PROCESO FINALIZADO")
    print("==============================")


if __name__ == "__main__":
    main()
```

```python
import os
import sys

inventario = {
1: {"nombre": "Martillo", "precio": 150000, "stock": 10},
2: {"nombre": "Destornillador", "precio": 5000, "stock": 20},
3: {"nombre": "Taladro", "precio": 85000, "stock": 4},
4: {"nombre": "Llave inglesa", "precio": 12000, "stock": 15}
}

factura_actual = []

def limpiar_pantalla():
    os.system('cls' if os.name == 'nt' else 'clear')

def mostrar_encabezado():
    print(f"{'ID':<6}{'NOMBRE':<20}{'PRECIO':<10}{'STOCK':<8}")
    print('-'*48)

def mostrar_inventario():
    print("\n *************INVENTARIO COMPLETO*************")
    if not inventario:
        print("El inventario esta vacio.")
    else:
        mostrar_encabezado()
        for id_prod, datos in inventario.items():
            print(f"{id_prod:<6}{datos['nombre']:<20}{datos['precio']:<9.2f}{datos['stock']:<8}")
        print('-'*48)

def buscar_producto_nombre():
    print("\n *************BUSCAR POR NOMBRE*************")
    termino = input("Ingrese el nombre del producto a buscar: ").strip().lower()
    encontrados = False
    print("Resultados de la busqueda: ")
    mostrar_encabezado()
    for id_prod, datos in inventario.items():
        if termino in datos['nombre'].lower():
            print(f"{id_prod:<6}{datos['nombre']:<20}{datos['precio']:<9.2f}{datos['stock']:<8}")
            encontrados = True
    if not encontrados:
        print("No se encontraron productos con ese nombre.")
    print('-'*48)

def agregar_producto():
    print("\n *************AGREGAR NUEVO PRODUCTO*************")
    id_prod_input = input("Ingrese un ID unico: ").strip()
    try:
        id_prod = int(id_prod_input)
        if id_prod in inventario:
            print("ERROR: Ya existe ese ID.")
        else:
            nombre = input("Nombre del producto: ").strip()
            precio = float(input("Precio del producto: "))
            stock = int(input("Stock del producto: "))
            inventario[id_prod] = {
                "nombre": nombre,
                "precio": precio,
                "stock": stock
            }
            print(f"Producto '{nombre}' creado con exito" )
    except ValueError:
        print("ERROR: El ID, precio y stock deben ser numeros validos.")

def modificar_producto():
    print("\n *************MODIFICAR PRODUCTO*************")
    mostrar_inventario()
    id_prod_input = input("Ingrese el ID del producto a modificar: ").strip()
    try:
        id_prod = int(id_prod_input)
        if id_prod in inventario:
            prod = inventario[id_prod]
            print(f"\nProducto seleccionado: {prod['nombre']}, Stock: {prod['stock']}, Precio: ${prod['precio']:.2f}")
            print("1. Agregar unidades")
            print("2. Cambiar precio")
            print("3. Cancelar")
            op_mod = input("Seleccione una opcion: ")

            match op_mod:
                case "1":
                    cantidad = int(input("¿Cuantas unidades desea agregar?: "))
                    if cantidad > 0:
                        prod['stock'] += cantidad
                        print(f"Stock actualizado. Nuevo total: {prod['stock']} unidades.")
                    else:
                        print("La cantidad debe ser positiva")

                case "2":
                    nuevo_precio = float(input("Ingrese el nuevo precio: "))
                    if nuevo_precio >= 0:
                        prod['precio'] = nuevo_precio
                        print(f"Precio actualizado a: ${prod['precio']:.2f}")
                    else:
                        print("El precio no puede ser negativo")

                case "3":
                    print("Operacion cancelada.")

                case _:
                    print("Opcion no valida.")
        else:
            print("El producto no fue encontrado.")

    except ValueError:
        print("ERROR: Entrada invalida. Debe ingresar numeros correctos.")

def eliminar_producto():
    print("\n *************ELIMINAR PRODUCTO*************")
    mostrar_inventario()
    id_prod = input("Ingrese el ID del producto a eliminar: ").strip()

    try:
        id_prod = int(id_prod)
        if id_prod in inventario:
            print(f"Va a eliminar: {inventario[id_prod]['nombre']}")
            confirmacion = input("¿Está seguro que desea eliminar el producto? (s/n): ").strip().lower()

            if confirmacion == 's':
                eliminado = inventario.pop(id_prod)
                print(f"Producto '{eliminado['nombre']}' eliminado con exito.")
            else:
                print("Operacion cancelada.")
        else:
            print("ID no encontrado.")
    except ValueError:
        print("ERROR: ID invalido.")


def vender_producto():
    print("\n *************VENDER PRODUCTO*************")
    mostrar_inventario()
    buscar = input("¿Desea buscar el ID por nombre? (s/n): ").lower()
    if buscar == 's':
        buscar_producto_nombre()

    id_prod_str = input("Ingrese el ID del producto a vender: ").strip()
    try:
        id_prod = int(id_prod_str)
        if id_prod in inventario:
            prod = inventario[id_prod]
            print(f"producto: {prod['nombre']} Disponibles: {prod['stock']}")
            cantidad_str = input("Cantidad a vender: ")

            try:
                cantidad = int(cantidad_str)
                if 0 < cantidad <= prod['stock']:
                    subtotal = cantidad * prod['precio']
                    prod['stock'] -= cantidad
                    factura_actual.append({
                        'nombre': prod['nombre'],
                        'precio': prod['precio'],
                        'cantidad': cantidad,
                        'subtotal': subtotal
                    })
                    print(f"Agregado al carrito: ${subtotal:.2f}")
                else:
                    print("Cantidad invalida o stock insuficiente")

            except ValueError:
                print("ERROR: La cantidad debe ser un número entero válido.")
        else:
            print("ERROR: Producto no encontrado.")

    except ValueError:
        print("ERROR:")


def generar_factura():
    print("\n" + "=" * 40)
    print("FERRETERIA 'LE FALTA UN TORNILLO'")
    print("*       FACTURA DE VENTA       *")
    print("=" * 40)

    if not factura_actual:
        print("No hay items en la factura actual.")
    else:
        total = 0
        print(f"{'CANT':<5}{'ARTICULO':<20}{'UNIDAD':<8}{'SUBTOTAL':<10}")
        print("-" * 45)

        for item in factura_actual:
            print(f"{item['cantidad']:<5}{item['nombre']:<20}{item['precio']:<7.2f}{item['subtotal']:<9.2f}")
            total += item['subtotal']

        print("-" * 45)
        print(f"total a pagar: ${total:.2f}")
        print("=" * 40)

        if input("\n ¿Confirmar el pago y cerrar venta? (s/n): ").lower() == 's':
            factura_actual.clear()
            print("Venta finalizada. Gracias por su compra.")


def main():
    while True:
        print("=============SISTEMA DE GESTION DE INVENTARIO=============")
        print("1. Mostrar inventario completo")
        print("2. Buscar inventario por nombre")
        print("3. Vender (Agregar a Factura)")
        print("4. Generar factura / cerrar venta")
        print("=" * 30)
        print("5. Agregar nuevo producto")
        print("6. Modificar producto (Stock/precio)")
        print("7. Eliminar producto")
        print("8. Salir")

        opcion = input("Seleccione una opcion: ")

        match opcion:
            case "1":
                mostrar_inventario()
            case "2":
                buscar_producto_nombre()
            case "3":
                vender_producto()
            case "4":
                generar_factura()
            case "5":
                agregar_producto()
            case "6":
                modificar_producto()
            case "7":
                eliminar_producto()
            case "8":
                print("Saliendo del programa...")
                break

            case _:
                print("Opcion no valida")

        input("Presione enter para continuar...")
        limpiar_pantalla()


if __name__ == "__main__":
    if sys.version_info < (3,10):
        print("Error de python")
    else:
        main()
```

```python
import os
import sys

estudiantes = {
1: {"nombre": "Juan Perez", "grado": "10A", "asistencias": 10, "faltas": 2, "correo": "juan@mail.com", "telefono": "3001234567"},
2: {"nombre": "Maria Lopez", "grado": "9B", "asistencias": 12, "faltas": 1, "correo": "maria@mail.com", "telefono": "3009876543"},
3: {"nombre": "Carlos Gomez", "grado": "11A", "asistencias": 8, "faltas": 3, "correo": "carlos@mail.com", "telefono": "3014567890"},
4: {"nombre": "Ana Torres", "grado": "8C", "asistencias": 15, "faltas": 0, "correo": "ana@mail.com", "telefono": "3026547891"}
}

registro_asistencia = []

def limpiar_pantalla():
    os.system('cls' if os.name == 'nt' else 'clear')

def mostrar_encabezado():
    print(f"{'ID':<4} {'NOMBRE':<20} {'GRADO':<6} {'ASIST':<6} {'FALTAS':<7} {'CORREO':<25} {'TEL':<12}")
    print('-'*90)

def mostrar_estudiantes():
    print("\n *************LISTA DE ESTUDIANTES*************")
    if not estudiantes:
        print("No hay estudiantes.")
    else:
        mostrar_encabezado()
        for id_est, datos in estudiantes.items():
           print(f"{id_est:<4} {datos['nombre']:<20} {datos['grado']:<6} {datos['asistencias']:<6} {datos['faltas']:<7} {datos['correo']:<25} {datos['telefono']:<12}")
        print('-'*80)

def buscar_estudiante():
    print("\n *************BUSCAR ESTUDIANTE*************")
    termino = input("Nombre: ").strip().lower()
    encontrados = False
    mostrar_encabezado()
    for id_est, datos in estudiantes.items():
        if termino in datos['nombre'].lower():
            print(f"{id_est:<5}{datos['nombre']:<20}{datos['grado']:<8}{datos['asistencias']:<6}{datos['faltas']:<6}{datos['correo']:<20}{datos['telefono']:<12}")
            encontrados = True
    if not encontrados:
        print("No encontrado")
    print('-'*80)

def agregar_estudiante():
    print("\n *************AGREGAR ESTUDIANTE*************")
    try:
        id_est = int(input("ID: "))
        if id_est in estudiantes:
            print("Ya existe")
        else:
            nombre = input("Nombre: ")
            grado = input("Grado: ")
            correo = input("Correo: ")
            telefono = input("Telefono: ")

            estudiantes[id_est] = {
                "nombre": nombre,
                "grado": grado,
                "asistencias": 0,
                "faltas": 0,
                "correo": correo,
                "telefono": telefono
            }
            print("Agregado")
    except ValueError:
        print("ERROR")

def modificar_estudiante():
    print("\n *************MODIFICAR*************")
    mostrar_estudiantes()
    try:
        id_est = int(input("ID: "))
        if id_est in estudiantes:
            est = estudiantes[id_est]
            print("1. Asistencia")
            print("2. Falta")
            print("3. Modificar correo")
            print("4. Modificar telefono")
            op = input("Opcion: ")

            match op:
                case "1":
                    est['asistencias'] += 1
                    print("Asistencia registrada")
                case "2":
                    est['faltas'] += 1
                    print("Falta registrada")
                case "3":
                    nuevo_correo = input("Nuevo correo: ")
                    est['correo'] = nuevo_correo
                    print("Correo actualizado")
                case "4":
                    nuevo_tel = input("Nuevo telefono: ")
                    est['telefono'] = nuevo_tel
                    print("Telefono actualizado")
                case _:
                    print("Invalido")
        else:
            print("No existe")
    except ValueError:
        print("ERROR")

def eliminar_estudiante():
    print("\n *************ELIMINAR*************")
    mostrar_estudiantes()
    try:
        id_est = int(input("ID: "))
        if id_est in estudiantes:
            if input("Seguro? (s/n): ").lower() == 's':
                eliminado = estudiantes.pop(id_est)
                print(f"{eliminado['nombre']} eliminado")
        else:
            print("No encontrado")
    except ValueError:
        print("ERROR")

def registrar_asistencia():
    print("\n *************REGISTRAR*************")
    mostrar_estudiantes()
    try:
        id_est = int(input("ID: "))
        if id_est in estudiantes:
            est = estudiantes[id_est]
            op = input("1.Asistencia 2.Falta: ")

            if op == "1":
                est['asistencias'] += 1
                registro_asistencia.append({
                    "nombre": est['nombre'],
                    "grado": est['grado'],
                    "tipo": "Asistencia"
                })
                print("OK")
            elif op == "2":
                est['faltas'] += 1
                registro_asistencia.append({
                    "nombre": est['nombre'],
                    "grado": est['grado'],
                    "tipo": "Falta"
                })
                print("OK")
            else:
                print("Invalido")
        else:
            print("No existe")
    except ValueError:
        print("ERROR")

def ver_reporte():
    print("\n" + "="*40)
    print("REPORTE DE ASISTENCIA")
    print("="*40)

    if not registro_asistencia:
        print("Sin registros")
    else:
        for r in registro_asistencia:
            print(f"{r['nombre']} ({r['grado']}) - {r['tipo']}")

        if input("Limpiar? (s/n): ") == 's':
            registro_asistencia.clear()

def main():
    while True:
        print("===== CONTROL DE ASISTENCIAS =====")
        print("1. Ver estudiantes")
        print("2. Buscar")
        print("3. Registrar")
        print("4. Reporte")
        print("5. Agregar")
        print("6. Modificar")
        print("7. Eliminar")
        print("8. Salir")

        op = input("Opcion: ")

        match op:
            case "1": mostrar_estudiantes()
            case "2": buscar_estudiante()
            case "3": registrar_asistencia()
            case "4": ver_reporte()
            case "5": agregar_estudiante()
            case "6": modificar_estudiante()
            case "7": eliminar_estudiante()
            case "8":
                break
            case _:
                print("Invalido")

        input("ENTER...")
        limpiar_pantalla()

if __name__ == "__main__":
    if sys.version_info < (3,10):
        print("Error")
    else:
        main()
```
